```python
import mne
from MEG_Tools import MEG
from MEG_Tools import Graph
import matplotlib
import numpy as np
#matplotlib.use('Qt5Agg') #Allows Charts to pop-up and use interactive features
import matplotlib.pyplot as plt
```

```python
meg = MEG('case_2225_with_spike_dipoles_sleep_2.mat') #must have data file in this directory
```

```python
#Sample Peaks - With Spike
#This plot uses native plotting (from MEG object) and not MNE Plotting
#get_peaks uses the get_gaussian_smoothed() function shown in the plot below

the_peaks = (meg.get_peaks("MEG0242",51.25,51.75,smoothing=7,plot=True))
(biggest_peak_index, biggest_peak, biggest_peak_width) = meg.max_peak("MEG0242",51.25,51.75)
(biggest_change_index, biggest_change) = meg.max_change("MEG0242",51.25,51.75)

print("Biggest peak -> location "+str(biggest_peak_index)+" : "+str(biggest_peak)) #reminder indeces start at 0
print("Biggest change -> location "+str(biggest_change_index)+" : "+str(biggest_change)) #reminder indeces start at 0
print("Width of biggest peak -> location "+str(biggest_peak_index)+" : "+str(biggest_peak_width))

meg.plot_channel_slice("MEG0242",51.25,51.75)
```

```python
#Sample Peak - No Spike
meg.get_peaks("MEG0113",0.0, 0.5,smoothing=7, plot=True)
(biggest_peak_index, biggest_peak, biggest_width) = meg.max_peak("MEG0113",0.0, 0.5  )
(biggest_change_index, biggest_change) = meg.max_change("MEG0113",0.0, 0.5  )

print("Biggest peak -> location "+str(biggest_peak_index)+" : "+str(biggest_peak)) #reminder indeces start at 0
print("Biggest change -> location "+str(biggest_change_index)+" : "+str(biggest_change)) #reminder indeces start at 0
print("Biggest width -> location "+str(biggest_peak_index)+" : "+str(biggest_width)) #reminder indeces start at 0

meg.plot_channel_slice("MEG0113",0.0, 0.5)
```

```python
#Filtering
meg.plot_channel_slice("MEG0113",1.0, 3.0)
filtered = meg.copy()
filtered.filter_data(lowcut=3.0,highcut=35.0) #these are the defaults, but are shown for clarity
filtered.plot_channel_slice("MEG0113",1.0, 3.0)
filtered = meg.get_peaks("MEG0113",1.0, 3.0, plot=True)
print('')
print("with bandstop->what was removed")
filtered_2 = meg.copy()
filtered_2.filter_data(lowcut=3.0,highcut=35.0,type="bandstop") #these are the defaults, but are shown for clarity
filtered_2.plot_channel_slice( "MEG0113",1.0, 3.0)

```
